```python
import pandas as pd
import numpy as np

file_paths = [
    "./dataset_rs/RS_Umum_Daerah_Syarifah_Ambainii_Rato_Ebu_Report_TB_06_sulisrs_Januari-Desember_2020.xls",
    "./dataset_rs/RS_Umum_Daerah_Syarifah_Ambainii_Rato_Ebu_Report_TB_06_sulisrs_Januari-Desember_2021.xls",
    "./dataset_rs/RS_Umum_Daerah_Syarifah_Ambainii_Rato_Ebu_Report_TB_06_sulisrs_Januari-Desember_2022.xls",
    "./dataset_rs/RS_Umum_Daerah_Syarifah_Ambainii_Rato_Ebu_Report_TB_06_sulisrs_Januari-Desember_2023.xls",
    "./dataset_rs/RS_Umum_Daerah_Syarifah_Ambainii_Rato_Ebu_Report_TB_06_sulisrs_Januari-Desember_2024.xls"
]
    
gabungan_df_data = []

for path in file_paths:
    data_frame = pd.read_excel(
    path, 
    skiprows=[x for x in range(0, 15)], 
    usecols="D,F,H,I,X:AA,AC,AD,AF", 
    names= ['Tanggal_Daftar', 'NIK', 'Umur', 'Jenis_Kelamin', 'Lokasi_Anatomi', 'Hasil_Pemeriksaan_Foto_Toraks', 'Status_HIV', 'Riwayat_Diabetes_Melitus', 'Hasil_Pemeriksaan_Diagnosis_TBC', 'Jenis_Pemeriksaan_Diagnosis', 'Tipe_Diagnosis'] ,
    dtype={
        'Tanggal_Daftar': str,
        'NIK': str,
        'Umur': np.int32,
        'Jenis_Kelamin': str,
        'Lokasi_Anatomi': str,
        'Hasil_Pemeriksaan': str,
        'Hasil_Pemeriksaan_Foto_Toraks': str,
        'Status_HIV': str,
        'Riwayat_Diabetes_Melitus': str,
        'Hasil_Pemeriksaan_Diagnosis_TBC': str,
        'Tipe_Diagnosis': str
    })
    gabungan_df_data.append(data_frame)

gabungan_df_data = pd.concat(gabungan_df_data)

# gabungan_df_data.to_excel("./test.xlsx")

total_nik = gabungan_df_data['NIK'].value_counts()


gabungan_df_data
```
